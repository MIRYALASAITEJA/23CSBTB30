```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
N = 1000
```

```python
defective_counts = np.arange(0, 9)
probabilities = np.array([0.10, 0.15, 0.20, 0.18, 0.12, 0.10, 0.08, 0.05, 0.02])
```

```python
defective_data = np.random.choice(defective_counts, size=N, p=probabilities)
```

```python
unique, counts = np.unique(defective_data, return_counts=True)
pmf = counts / N
```

```python
cdf = np.cumsum(pmf)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=unique, y=pmf, color='skyblue', edgecolor='black')
plt.xlabel("Number of Defective Items per Batch")
plt.ylabel("Probability (PMF)")
plt.title("Probability Mass Function (PMF) of Defective Items per Batch")
plt.xticks(unique)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=unique, y=cdf, marker='o', color='red', linestyle='-')
plt.xlabel("Number of Defective Items per Batch")
plt.ylabel("Cumulative Probability (CDF)")
plt.title("Cumulative Distribution Function (CDF) of Defective Items per Batch")
plt.xticks(unique)
plt.ylim(0, 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import scipy.stats as stats
```

```python
N = 100
p = 0.03
```

```python
prob_5_defective = stats.binom.pmf(5, N, p)
```

```python
prob_at_least_8 = 1 - stats.binom.cdf(7, N, p)
```

```python
expected_defective = N * p
```

```python
print(f"Probability of exactly 5 defective bulbs: {prob_5_defective:.4f}")
print(f"Probability of at least 8 defective bulbs: {prob_at_least_8:.4f}")
print(f"Expected number of defective bulbs: {expected_defective:.4f}")
```
